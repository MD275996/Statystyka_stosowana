```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sb
```

### Zad 7

```python
def dyst_z(x):
    return (1-np.exp(-x) - np.exp(-2*x) + np.exp(-3*x))
def dyst_w(x):
    return 1-np.exp(-3*x)
```

```python
def proba_z_i_w(n):
    zs = np.zeros(n)
    ws = np.zeros(n)
    for i in range(n):
        x = np.random.exponential(1/2)
        y = np.random.exponential(1)
        zs[i] = max(x,y)
        ws[i] = min(x,y)
    return zs, ws
```

```python
n = 10000
zs,ws = proba_z_i_w(n)
sb.ecdfplot(zs, color="blue",label = "z empiryczna")
sb.ecdfplot(ws, color="orange", label = "w empiryczna")
rng = np.linspace(0,20,n)
plt.plot(rng, dyst_z(rng), color = "purple", label="z teoretyczna")
plt.plot(rng, dyst_w(rng), color = "red", label="w teoretyczna")
plt.legend()
```

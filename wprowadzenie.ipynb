```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
import seaborn as sb
```

```python
l=5
def gestosc(x):
    if x.all()>0:
        return(l*np.exp(-l*x))
    else:
        return 0

def dyst(x):
    if x.any() <= 0:
        return 0
    elif x.any() > 0:
        return 1-np.exp(-l*x)

x_lim = np.linspace(-1,1,100)
y_lim = gestosc(x_lim)
y_lim2 = dyst(x_lim)

plt.plot(x_lim, y_lim)
plt.plot(x_lim, y_lim2)
# plt.plot(x_lim, sps.expon.cdf(x_lim,0,1/l)) #dyst
# plt.plot(x_lim, sps.expon.pdf(x_lim,0,1/l)) #gestość
```

```python
x= np.random.normal(0,1,1000) #generowanie zmiennych losowych o rozkładzie normalnym
# plt.scatter(range(0,1000),x)
rng = np.linspace(-4,4,1000)
#policzyc empiryczną dystrybuantę
dystr = sb.ecdfplot(x)
y = sps.norm.cdf(rng,0,1)
plt.plot(rng,y)
```

```python
#gęstości
y2 = sps.norm.pdf(rng,0,1)
plt.plot(rng,y2,label="analityczna")
sb.histplot(x,stat='density',label="empiryczna")
sb.kdeplot(x, label="empiryczna")
plt.legend()
```
